### **Background:**<br>

Organizations are competing in an unprecedented way in today’s global economy. Over the past two decades, we are seeing a shift in the perceived value of an organization. In the past, consultants and academics believe in looking at _Property and Equipment_ as the two most important determinants of value.  With the economy transitioning towards a knowledge based economy, the most valuable asset of an organization is now **_Human Capital_**. <br> <br>
The rise of start-ups driven by talented people building disruptive products or services has proven to us that young companies which are typically small in size with strong human capital can compete at the same level as the giant corporations. <br> <br>
The transition from _Capitalism_ to _Talentism_ is highlighted in 2014 by the founder of the World Economic Forum, Professor Klaus Schwab, who stated: <br><br> _“Capital is being superseded by creativity and the ability to innovate — and therefore by human talents — as the most important factors of production. If talent is becoming the decisive competitive factor, we can be confident in stating that capitalism is being replaced by ‘talentism.’ Just as capital replaced manual trades during the process of industrialization, capital is now giving way to human talent.”_ <br> <br>
For organization to thrive in today’s global competition calls for significant investment in Human Capital. Human Resource function of the organization holds one of the important keys to business success. Analysing workforce data over time provides valuable insights to people performance, talent management, employee engagement, undesirable attrition, and return of investment. <br> <br>
Against this backdrop, **this project aims to study key features leading to attrition of high performing employees**. The predictive model built will be used to indicate if the company is at risk of losing a talent and we can suggest possible remedies to prevent that by addressing the key features as identified by the model.

(i) Define a function for ploting stacked bar chart

```python
def plot_stacked_bar_chart( feature1, feature2, data, transpose = True, ax=None, figsize = (8,8)):
    df = data[[feature1, feature2]].copy()
    df1 = df.groupby(feature1)[feature2].value_counts()
    df1 = df1.reset_index(level=[feature1])
    split = int(len(df1.index)/2)
  
    no_df1 = df1.iloc[0:split]
    no_df1 = no_df1.rename(columns={feature2: feature1+': ' + no_df1.iloc[0][feature1]}).drop([feature1], axis=1)
    yes_df1 = df1.iloc[split:len(df1.index)]
    yes_df1 = yes_df1.rename(columns={feature2: feature1+': ' + yes_df1.iloc[0][feature1]}).drop([feature1], axis=1)
    df2 = pd.concat( [no_df1,yes_df1], axis=1)
    df2 = df2.sort_index()
    if transpose:
        if ax:
            df2.T.plot(kind='bar', stacked=True, ax=ax, figsize=figsize)
        else:
            df2.T.plot(kind='bar', stacked=True, figsize=figsize)
    else:
        if ax:
            df2.plot(kind='bar', stacked=True, ax=ax, figsize=figsize)
        else:
            df2.plot(kind='bar', stacked=True, figsize=figsize)
            
    return df2
            
```

###  4 (A) Decision Tree: predicting attrition using all features
(i) Perform Grid Search-DecisionTree modelling -- identify the Score and the parameters <br>
(ii) Perform 10-fold Cross-Validation -- identify the Score

```python
#train/test split
from sklearn.model_selection import train_test_split
X=HrData_clean
y=HrData.attrition_num

(training_inputs,
 testing_inputs,
 training_classes,
 testing_classes) = train_test_split(X, y, test_size=0.25, random_state=19)
```

```python
#Run decision tree for all features.
from sklearn.tree import DecisionTreeClassifier

# Create the classifier
tree_class = DecisionTreeClassifier(random_state=19)

# Train the classifier on the training set
tree_class.fit(training_inputs, training_classes)

# Validate the classifier on the testing set using classification accuracy - decision tree score is accuracy
tree_class.score(testing_inputs, testing_classes)
```

```python
import sklearn.tree as tree
from sklearn.externals.six import StringIO

with open('HR_DTC.dot', 'w') as out_file:
    out_file = tree.export_graphviz(tree_class, out_file=out_file,  feature_names=X.columns)
```

```python
#Examine model accuracies for 1000 train/test splitsand 1000 DT
plt.rcParams['figure.figsize'] = (7, 4)
plt.rcParams['font.size'] = 14


model_accuracies = []

for repetition in range(1000):
    (training_inputs,
     testing_inputs,
     training_classes,
     testing_classes) = train_test_split(X, y, test_size=0.25)
    
    tree_class = DecisionTreeClassifier(random_state=19)
    tree_class.fit(training_inputs, training_classes)
    classifier_accuracy = tree_class.score(testing_inputs, testing_classes)
    model_accuracies.append(classifier_accuracy)
    
ma_plot=sns.distplot(model_accuracies)
ma_plot.set(xlabel='Model Accuracies', ylabel='Count')
plt.show()
```

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score

cross_validation = StratifiedKFold(n_splits=10, random_state = 19, shuffle=True)
cross_validation
```

```python
tree_class = DecisionTreeClassifier(random_state=19)

# cross_val_score returns a list of the scores, which we can visualize
# to get a reasonable estimate of our classifier's performance
cv_scores = cross_val_score(tree_class, X, y, cv=cross_validation)
print ('CV scores for n_folds=10:{}'.format(cv_scores))
print ('\nAverage CV_scores: {}'.format(np.mean(cv_scores)))
```

```python
#Search for optimal max_depth and max_features

from sklearn.model_selection import GridSearchCV

tree_class = DecisionTreeClassifier(random_state=19)

parameter_grid = {'criterion': ['gini'],
                  'splitter': ['best', 'random'],
                  'max_depth': range(1,11),
                  'max_features': range(1,int(len(X.columns)/2))}

#cross_validation = StratifiedKFold(y, n_folds=10)

grid_search_dt = GridSearchCV(tree_class,
                           param_grid=parameter_grid,
                           cv=cross_validation,
                             n_jobs = -1)

grid_search_dt.fit(X, y)
print('Best score: {}'.format(grid_search_dt.best_score_))
DT_allimp_accu=grid_search_dt.best_score_
print('Best parameters: {}'.format(grid_search_dt.best_params_))
```

```python
tree_class_best = grid_search_dt.best_estimator_
tree_class_best
```

```python
scores = cross_val_score(tree_class_best, X, y, cv=cross_validation)
DTscore_allfeat=np.mean(scores)

print ("Average of 10-fold CV score: {}".format(np.mean(scores)))
print('Best parameters: {}'.format(tree_class_best.get_params()))

pd.DataFrame({'feature':X.columns, 'importance':tree_class_best.feature_importances_}).sort_values(by='importance')
```

### Step 5: Confusion Matrix

```python
# Create function for plotting Confusion Matrix 
# Refence: http://scikit-learn.org/stable/auto_examples/model_selection/plot_confusion_matrix.html
import itertools
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)
    
    
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
   
```

3) This scenario can occur when we have imbalanced dataset that causes the predictive models to fit very well for the majority class ("No-Attrition") but fairly poor for the minority class ("Attrition"). <br>


<a id="Step 6"></a>
### Step 6: Solving Imbalanced Class <br>

The dataset has highly imbalanced classes (16% "Attrition" and 84% "No-Attrition") that is not addressed during the preprocessing. One way to solve this issue is to use "down-sampling" technique which can be achieved by reducing the number of samples in the majority class to obtain roughly equal number of data across the classes before training. Another option to solve the imbalanced class is  by setting the "class_weight" argument in the predictive model to "balanced" to ensure  a balanced mix of each class in training. The "balanced" setting will automatically weigh classes inversely proportional to the frequency of occurence in the dataset.
