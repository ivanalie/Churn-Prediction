### Part 3: Calculate the average price of an order.

Hint: Examine the data to see if the `quantity` column is relevant to this calculation.

Hint: Think carefully about the simplest way to do this!

```python
total=0
n=0
quantity=0

for no,line in enumerate(data):
    for x,y in enumerate(line):
        if x==1:
            quantity=float(y)
            n=n+float(y)
        if x==4:
            price=float(y.strip('$'))
            total=total+(price*quantity)

last_order_id=int(data[-1][0])

print ("The last order = %s. The total price: $%s" %(last_order_id,total)) 
print ("The average price of an order = total price/last order: $%s" %(total/last_order_id))
```

### Part 4: Create a list (or set) named `unique_sodas` containing all of unique sodas and soft drinks that Chipotle sells.

Note: Just look for `'Canned Soda'` and `'Canned Soft Drink'`, and ignore other drinks like `'Izze'`.

```python
unique_sodas=[]
for line in data:
    for x,y in enumerate(line):
        if x==2:
            if 'Canned Soda' in y or 'Canned Soft Drink' in y:
                #print (y, line[x+1][0:3])
                if line[x+1][1:-1] not in unique_sodas:
                    unique_sodas.append(line[x+1][1:-1])

print (unique_sodas)
```

### Part 5: Calculate the average number of toppings per burrito.

Note: Let's ignore the `quantity` column to simplify this task.

Hint: Think carefully about the easiest way to count the number of toppings!


```python
total_burrito=0
total_toppings=0
for line in data:
    for x,y in enumerate(line):
        if x==2:
            if 'burrito' in y.lower():
                total_burrito+=1
                #split line[x+1] to two list: first list => sauce; second list => toppings
                cleanline = line[x+1].replace('[','').replace(']','')
                sauce = cleanline.split(',')[0]
                toppings = cleanline.split(',')[1:]
                total_toppings += len(toppings)
                #print (y,sauce, toppings, len(toppings))
                
print("The average number of toppings per burrito: {0}".format(float(total_toppings)/total_burrito))
```
